```python
import numpy as np
import keras
import matplotlib.pyplot as plt
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(x_train, y_train), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
def column_wise_encode(img):
    """
    Scan the given image column wise (left to  right) and expand pixel values to 1d array

    Parameters:
        img: np.ndarray
            Given image
    
    Returns:
        out: list
            Scanned pixel values
    """
    H, W = img.shape
    out = []
    
    for w in range(W):
        for h in range(H):
            out.append(img[h, w])
    
    return out
```

```python
test = x_train[17, :, :]
plt.imshow(test)
```

```python
c = column_wise_encode(test)
c_img = np.array(c).reshape(28, 28).astype(np.uint8).T
plt.imshow(c_img)
```
