ラベルごとの各フィルターの効果を可視化していきます.とりあえず，fashionmnistの画素をX_train,ラベルをy_trainとして読み込みます。

```python
import numpy as np
import matplotlib.pyplot as plt
from mnistreader import load_mnist
X_train, y_train = load_mnist('', kind='train')
```

ラベルごとに画素の値を分けていきます。画素の値とラベルをくっつけた後，ラベルを基準に降順にソートし，上から取り出していきます。

```python
pixelandlavels = np.hstack([X_train, y_train.reshape(60000,1)])
pixelandlavels_sorted = pixelandlavels[np.argsort(pixelandlavels[:, -1])]
pixel_label0 = pixelandlavels_sorted[0:6000]
pixel_label1 = pixelandlavels_sorted[6000:12000]
pixel_label2 = pixelandlavels_sorted[12000:18000]
pixel_label3 = pixelandlavels_sorted[18000:24000]
pixel_label4 = pixelandlavels_sorted[24000:30000]
pixel_label5 = pixelandlavels_sorted[30000:36000]
pixel_label6 = pixelandlavels_sorted[36000:42000]
pixel_label7 = pixelandlavels_sorted[42000:48000]
pixel_label8 = pixelandlavels_sorted[48000:54000]
pixel_label9 = pixelandlavels_sorted[54000:60000]
```

ラベルごとの輝度値の分布は以下の通りです。ラベル0,1,..9の順で示します。

```python
plt.hist(np.ravel(pixel_label0),bins=255 ,rwidth=0.8, range=(0, 255))
plt.title("ravel0")
plt.show()
```
